```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
from nltk.corpus import stopwords
import seaborn as sns
import nltk

# Download stopwords
nltk.download('stopwords')

# Load the dataset
train_df = pd.read_csv("train.csv")

# Select relevant columns
train_df = train_df[['question1', 'question2', 'is_duplicate']]

# Fill missing values with empty strings
train_df['question1'] = train_df['question1'].fillna('')
train_df['question2'] = train_df['question2'].fillna('')

# Define stopwords
stop_words = set(stopwords.words('english'))

# Function to preprocess text and count stop words
def count_stop_words(texts):
    stop_words_counts = Counter()
    for text in texts:
        words = text.lower().split()
        words = [word for word in words if word in stop_words]
        stop_words_counts.update(words)
    return stop_words_counts

# Combine question1 and question2 for each class
class_0_texts = pd.concat([train_df[train_df['is_duplicate'] == 0]['question1'],
                           train_df[train_df['is_duplicate'] == 0]['question2']])
class_1_texts = pd.concat([train_df[train_df['is_duplicate'] == 1]['question1'],
                           train_df[train_df['is_duplicate'] == 1]['question2']])

# Count stop words occurrence for each class
class_0_stop_words_counts = count_stop_words(class_0_texts)
class_1_stop_words_counts = count_stop_words(class_1_texts)

# Get top 10 most common stop words for each class
top_n = 10
class_0_top_stop_words = class_0_stop_words_counts.most_common(top_n)
class_1_top_stop_words = class_1_stop_words_counts.most_common(top_n)

# Convert results to DataFrame for visualization
class_0_df = pd.DataFrame(class_0_top_stop_words, columns=['Stop Word', 'Count'])
class_1_df = pd.DataFrame(class_1_top_stop_words, columns=['Stop Word', 'Count'])

# Plotting with annotations
plt.figure(figsize=(12, 6))

# Plot for Non-duplicate (is_duplicate=0)
plt.subplot(1, 2, 1)
plot1 = sns.barplot(x='Stop Word', y='Count', data=class_0_df, palette='Blues_d')
plt.title('Top 10 Stop Words in Non-duplicate Questions (0)')
plt.xticks(rotation=45)
for index, row in class_0_df.iterrows():
    plot1.text(index, row['Count'] + 0.2, row['Count'], color='black', ha="center")

# Plot for Duplicate (is_duplicate=1)
plt.subplot(1, 2, 2)
plot2 = sns.barplot(x='Stop Word', y='Count', data=class_1_df, palette='Reds_d')
plt.title('Top 10 Stop Words in Duplicate Questions (1)')
plt.xticks(rotation=45)
for index, row in class_1_df.iterrows():
    plot2.text(index, row['Count'] + 0.2, row['Count'], color='black', ha="center")

plt.tight_layout()
plt.savefig('Top_10_Stop_Words_in_Duplicate_and_Nonduplicate_Questions.png', dpi=600, bbox_inches='tight')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
import string
import seaborn as sns

# Load the dataset
train_df = pd.read_csv("train.csv")

# Select relevant columns
train_df = train_df[['question1', 'question2']]

# Function to count punctuation marks
def count_punctuation(texts):
    punctuation_counts = Counter()
    for text in texts:
        for char in text:
            if char in string.punctuation:
                punctuation_counts[char] += 1
    return punctuation_counts

# Combine question1 and question2 into a single list for analysis
all_questions = pd.concat([train_df['question1'].fillna(''), train_df['question2'].fillna('')])

# Count punctuation occurrence
punctuation_counts = count_punctuation(all_questions)

# Get top 10 most common punctuation marks
top_n = 10
top_punctuation = punctuation_counts.most_common(top_n)

# Convert results to DataFrame for visualization
punctuation_df = pd.DataFrame(top_punctuation, columns=['Punctuation', 'Count'])

# Plotting with annotations
plt.figure(figsize=(10, 6))
plot = sns.barplot(x='Punctuation', y='Count', data=punctuation_df, palette='viridis')
plt.title('Top 10 Most Common Punctuation Marks in Quora Questions')
plt.xlabel('Punctuation')
plt.ylabel('Count')
plt.xticks(rotation=45)

# Add count annotations on top of each bar
for index, row in punctuation_df.iterrows():
    plot.text(index, row['Count'] + 0.2, row['Count'], color='black', ha="center")

plt.tight_layout()
plt.savefig('Top_10_Most_Common_Punctuation_Marks_in_Quora_Questions.png', dpi=600, bbox_inches='tight')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')

# Load the dataset
train_df = pd.read_csv("train.csv")

# Select relevant columns
train_df = train_df[['question1', 'question2', 'is_duplicate']]

# Fill missing values in questions with empty strings
train_df['question1'] = train_df['question1'].fillna('')
train_df['question2'] = train_df['question2'].fillna('')

# Define stopwords
stop_words = set(stopwords.words('english'))

# Function to remove stopwords from text
def remove_stopwords(text):
    words = text.lower().split()
    return ' '.join([word for word in words if word not in stop_words])

# Apply stopword removal to both question1 and question2
train_df['question1_cleaned'] = train_df['question1'].apply(remove_stopwords)
train_df['question2_cleaned'] = train_df['question2'].apply(remove_stopwords)

# Combine question1 and question2 into a single string for analysis
train_df['all_questions'] = train_df['question1_cleaned'] + " " + train_df['question2_cleaned']

# Separate questions by target class (is_duplicate = 0 and is_duplicate = 1)
class_0_texts = train_df[train_df['is_duplicate'] == 0]['all_questions']
class_1_texts = train_df[train_df['is_duplicate'] == 1]['all_questions']

# Convert to single strings for word cloud generation
class_0_text = ' '.join(class_0_texts)
class_1_text = ' '.join(class_1_texts)

# Split the texts into unique words for each class
unique_words_class_0 = set(class_0_text.split())
unique_words_class_1 = set(class_1_text.split())

# Find unique words in each class
exclusive_words_class_0 = unique_words_class_0 - unique_words_class_1
exclusive_words_class_1 = unique_words_class_1 - unique_words_class_0

# Function to visualize word cloud
def visualize_word_cloud(words, title):
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(' '.join(words))
    plt.figure(figsize=(10, 6))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.title(title)
    plt.axis('off')
    plt.savefig(f'{title}.png', dpi=600, bbox_inches='tight')
    plt.show()

# Visualize unique words for non-duplicate questions (class 0)
visualize_word_cloud(exclusive_words_class_0, 'Unique Words in Non-duplicate Questions (Class 0)')

# Visualize unique words for duplicate questions (class 1)
visualize_word_cloud(exclusive_words_class_1, 'Unique Words in Duplicate Questions (Class 1)')
```
